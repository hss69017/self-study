# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["font.family"] = "AppleGothic" # win: Malgun Gothic or HYGungSo-Bold(궁서체)
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title("꺾은선 그래프", fontdict = {"family":"AppleGothic", "size":20}) # 해당 객체에 대해서 font 개별 적용 가능 
```

```python
plt.plot(x, y)
plt.xlabel("X축")
plt.ylabel("Y축")
```

```python
plt.plot(x, y)
plt.xlabel("X축", color = "red")
plt.ylabel("Y축", color = "#00aa00")
```

```python
plt.plot(x, y)
plt.xlabel("X축", color = "red", loc = "right") # loc: left, center, right 글자 위치
plt.ylabel("Y축", color = "#00aa00", loc = "top") # loc: top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
