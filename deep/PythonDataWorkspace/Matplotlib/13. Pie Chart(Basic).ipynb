# 13. 원 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["font.family"] = "AppleGothic" # win: Malgun Gothic or HYGungSo-Bold(궁서체)
matplotlib.rcParams["font.size"] = 15
matplotlib.rcParams["axes.unicode_minus"] = False
```

```python
values = [30, 25, 20, 13, 10, 2] # 굳이 다 더해서 100일 필요는 없음
labels = ["Python", "Java", "Javascript", "C#", "C/C++", "ETC"]

plt.pie(values, labels = labels, autopct = "%.1f%%", startangle = 90, counterclock = False) 
# autopct: 비율을 소수점 첫번째 자리까지 %로 보여줌, 뒤에 붙은 %%는 뒤에 %%를 붙여라! [X.X%]
# startangle: 해당 각도부터 시작, 0도는 3시 방향
# counterclock: 기본값은 True, 반시계방향
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ["Python", "Java", "Javascript", "C#", "C/C++", "ETC"]
explode = [0.2, 0.1, 0, 0, 0, 0] # 각 value를 다른 value와 얼마나 띄울지

plt.pie(values, labels = labels, explode = explode)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ["Python", "Java", "Javascript", "C#", "C/C++", "ETC"]
explode = [0.05] * 6

plt.pie(values, labels = labels, explode = explode)
plt.show()
```

```python
plt.pie(values, labels = labels, explode = explode)
plt.legend()
plt.show()
```

```python
plt.pie(values, labels = labels, explode = explode)
plt.legend(loc = (1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels = labels, explode = explode)
plt.title("언어")
plt.legend(loc = (1.2, 0.3), title = "언어별 선호도")
plt.show()
```
